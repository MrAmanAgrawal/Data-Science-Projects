# Data PreProcessing

#### Split DataFrame in train and test

```python
# input variable
X = cancer_data.drop(['target'], axis = 1)
X.head(6)
```

```python
# output variable
y = cancer_data['target']
y.head(6)
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 2020)
```

```python
X_train
```

```python
X_test
```

```python
y_train.shape
```

```python
y_test.shape
```

# Feature Scaling

##### Converting different units and magnitude data in one unit.

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

# Now our dataset is ready for machine learning model

## Breast Cancer Detection Machine Learning Model Building

### We have clean data to build the Ml model. But which Machine learning algorithm is best for the data we have to find. The output is a categorical format so we will use supervised classification machine learning algorithms.But for this project i have used Support Vector Classifier


First, we need to import the required packages.

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

### first we will use support vector classifier with rbf kernel

```python
# support vector classifier without scaled data
from sklearn.svm import SVC
svc_classifier_rbf = SVC(kernel = 'rbf')
svc_classifier_rbf.fit(X_train,y_train)
```

```python
svc_classifier_rbf.score(X_test,y_test)
```

```python
# support vector classifier with scaled data
from sklearn.svm import SVC
svc_classifier_rbf_2 = SVC(kernel = 'rbf')
svc_classifier_rbf_2.fit(X_train_sc,y_train)
```

```python
svc_classifier_rbf_2.score(X_test_sc,y_test)
```

#####  we can see there is an improvement in accuracy after we scaled the model

```python
from sklearn.svm import SVC
svc_classifier_poly = SVC(kernel = 'poly',degree = 2)
svc_classifier_poly.fit(X_train_sc,y_train)
```

```python
svc_classifier_poly.score(X_test_sc,y_test)
```

## we can see that svm with rbf kernel is more accurate , hence we will use the same for our prediction
